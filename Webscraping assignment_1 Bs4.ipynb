# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.imdb.com/chart/top/"
```

```python
page=requests.get(url)
page
```

```python
#display the page source code
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify()) # remove empty spaces & new lines
```

```python
#scraped movie names
scraped_movies=soup.find_all('td',class_='titleColumn')
scraped_movies
```

```python
#parse movie names
movies=[]
for movie in scraped_movies:
    movie =movie.get_text().replace('\n',"") #replacing blank & \n
    movie=movie.strip(" ")
    movies.append(movie)
movies    
```

```python
#scrap rating for movies
scraped_ratings=soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
#parse ratings
ratings=[]
for rating in scraped_ratings:
    rating= rating.get_text().replace('\n','')
    ratings.append(rating)
ratings    
```

```python
#scrap year for movies
scraped_years=soup.find_all('span',class_='secondaryInfo')
scraped_years
```

```python
#parse years
years=[]
for year in scraped_years:
    year=year.get_text().replace('\n',' ')
    years.append(year)
years 
```

```python
data=pd.DataFrame()
data['Movie names']=movies
data['Ratings']=ratings
data['year of release']=years
data.head(100)
```

```python
def odi_team(url):
    soup=BeautifulSoup(requests.get(url).content)#getting website content
    team=[]
    matches=[]
    points=[]
    ratings=[]
#taking 1st rank, matches,ponts, rating contents
    for i in soup.find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    for i in soup.find_all('td',class_="rankings-block__banner--matches"):
        matches.append(i.text)
    for i in soup.find_all('td',class_="rankings-block__banner--points"):
        points.append(i.text.strip())
    for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
        ratings.append(i.text)
        
#taking rest of the rank,matches,points,ratings
    table=[]
    for i in soup.find_all('tr',class_="table-body"):
        table.append(i.text.split())
    for i in table:
        matches.append(i[-3])
        points.append(i[-2])
        ratings.append(i[-1])
    data=pd.DataFrame({'Team':team[:10],'Matches':matches[:10],'Points':points[:10],'Ratings':ratings[:10]})
    data.index +=1
    return data
def odi(url):
    
    soup= BeautifulSoup(requests.get(url).content)
    player=[]
    team=[]
    ratings=[]
        
# take 1st rank from Banner
    for i in soup.find_all('div',class_="rankings-block__banner--name-large"):
        player.append(i.text)
    
    for i in soup.find_all('div',class_="rankings-block__banner--nationality"):
        team.append(i.text.strip())
    
    
    for i in soup.find_all('div',class_="rankings-block__banner--rating"):
        ratings.append(i.text)
 # take rest of the content from table
    for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
        player.append(i.text.strip())

    for i in soup.find_all('span',class_="table-body__logo-text"):
        team.append(i.text.strip())   
    
    for i in soup.find_all('td',class_="table-body__cell rating"):
        ratings.append(i.text.strip())  
        
    data= pd.DataFrame({"Player":player[:10],"Team":team[:10],'Ratings' :ratings[:10] })
    data.index +=1
    return data
```

7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
def coreyms(url):
    soup= BeautifulSoup(requests.get(url).content)  # Get website content

    data= requests.get('https://coreyms.com/')
    soup= BeautifulSoup(data.content)
    heading=[]
    date=[]
    content=[]
    video=[]

    for i in soup.find_all('h2',class_="entry-title"):
        heading.append(i.text)
        
    for i in soup.find_all('time',class_="entry-time"):
        date.append(i.text)
        
    for i in soup.find_all('div',class_="entry-content"):
        content.append(i.find('p').text)
        
    for i in soup.find_all('span', class_="embed-youtube"):
        video.append(i.find('iframe')['src'].strip())
        
    video.insert(4,'NA')
        
        
    data= pd.DataFrame({'Heading':heading,'Date': date, "Content": content,"Video Link": video  })  
    data.index +=1
    return data
    
```

```python
coreyms('https://coreyms.com/')
```

9) Write a python program to scrape mentioned details from dineout.co.in : 
    i) Restaurant name 
    ii) Cuisine
    iii) Location
    iv) Ratings 
    v)Image URL

```python
def dineout(url):
    #create enmpt list to store variables
    restro=[]
    cuisine=[]
    location=[]
    ratings=[]
    image_url=[]
    
    soup=BeautifulSoup(requests.get(url).content)
    
    for i in soup.find_all('div', class_="restnt-info cursor"):
        restro.append(i.text)
    
    for i in soup.find_all('span',class_="double-line-ellipsis" ):
        cuisine.append(i.text.split('|')[1])
    
    for i in soup.find_all('div',class_="restnt-loc ellipsis"  ):
        location.append(i.text)
    
    for i in soup.find_all('div',class_="restnt-rating rating-4"):
        ratings.append(i.text)
    
    for i in soup.find_all('img',class_="no-img" ):
        image_url.append(i['data-src'])
    
    data= pd.DataFrame({"Restaurant name":restro, "Cuisine":cuisine,"Location":location,"Ratings":ratings,"Image Url":image_url  })
    data.index+=1
    
    return data
```

```python
dineout( 'https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
def tshirt(url):
    data= requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
    soup= BeautifulSoup(data.content)
    product=[]
    price=[]
    image=[]
    for i in soup.find_all('div',class_="productCardBox"):
        product.append(i.find('h3').text)
        
    for i in soup.find_all('div',class_="productPriceBox clearfix"):
        price.append(i.find('span').text.replace("₹",""))
        
    for i in soup.find_all('div',class_="productCardImg false"):
        image.append(i.find('img')['src'])
        
        
    data= pd.DataFrame({'Product Name':product,'Price': price, "Image URL": image  })  
    data.index +=1
    
    return data
```

```python
tshirt( 'https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
```
